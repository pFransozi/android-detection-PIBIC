```python
import pandas as pd
```

```python
dframes = []

dframes.append((pd.read_json("./andropytool_output/MAL_OUTPUT_ANDROPY_1694726246.22.json"), 1))
```

```python
def is_to_drop(col):
  return not (col.startswith('Permissions') or col.startswith('Opcodes') or col.startswith('API calls'))
```

```python
class_col_name = "class"
class_col_pos = 0
static_analysis_col_name = "Static_analysis"
permissions_col_name = "Permissions"
dfs_concat = pd.DataFrame()
count = 0

for dframe, class_df in dframes:
  for col_apk in dframe.columns:

    df_static_analysis = pd.json_normalize(dframe[col_apk][static_analysis_col_name])
    df_static_analysis.drop([col for col in df_static_analysis.columns if is_to_drop(col)], axis=1, inplace=True)
    permissions = df_static_analysis.pop(permissions_col_name)[0]
    for perm in permissions:
      df_static_analysis[perm] = 1 if perm not in df_static_analysis else 1 + df_static_analysis[perm]

    df_static_analysis[class_col_name] = class_df
    dfs_concat = pd.concat([dfs_concat, df_static_analysis])

dfs_concat.fillna(0, inplace=True)
dfs_concat.reset_index(inplace=True)
dfs_concat.drop(["index"], axis=1, inplace=True)
```
